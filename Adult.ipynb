**Feature engineering**

```python
df = df.drop('workclass',axis=1).join(pd.get_dummies(df.workclass,prefix='wcls'))
df = df.drop('marital_status',axis=1).join(pd.get_dummies(df.marital_status,prefix='mstat'))
df = df.drop('occupation',axis=1).join(pd.get_dummies(df.occupation,prefix='occup'))
df = df.drop('relationship',axis=1).join(pd.get_dummies(df.relationship,prefix='rel'))
df = df.drop('race',axis=1).join(pd.get_dummies(df.race,prefix='race'))
df = df.drop('native_country',axis=1).join(pd.get_dummies(df.native_country,prefix='country'))
```

```python
df.sex = df.sex.map({'Female':0,'Male':1})
df.income = df.income.map({'<=50K':0,'>50K':1})
```

```python
correlation = df.corr().income.abs().copy()
sort_columns = correlation.sort_values(ascending=False)
num_cut_col = int(0.2*len(sort_columns))
drop_col = sort_columns.iloc[num_cut_col:].index
df_dropped = df.drop(drop_col,axis=1).copy()
```

```python
plt.figure(figsize=(15,10))
sns.heatmap(df_dropped.corr(),annot=True,cmap='coolwarm')
```

```python
#cut off 20% of df for the test
df_cut = int(len(df)*0.2) 
df_test = df.iloc[-df_cut:].copy()
df = df.iloc[:-df_cut]
```

**Decision Tree**

```python
target = 'income'
features = [col for col in df.columns if col != target]
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import KFold
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import accuracy_score
```

```python
kf = KFold(n_splits=10)
skf = StratifiedKFold(n_splits=10)
```

```python
avr_train_score = []
avr_valid_score = []

for i in range(2,150,5):
    
    train_score = []
    valid_score = []
    
    for train_ind,valid_ind in skf.split(df[features],df[target]):

        train = df.loc[train_ind].copy()
        valid = df.loc[valid_ind].copy()

        model = DecisionTreeClassifier(min_samples_leaf=i)
        model.fit(train[features],train[target])

        train['dectree_pred'] = model.predict(train[features])
        valid['dectree_pred'] = model.predict(valid[features])

        train_ac = accuracy_score(train[target],train['dectree_pred'])
        valid_ac = accuracy_score(valid[target],valid['dectree_pred'])

        train_score.append(train_ac)
        valid_score.append(valid_ac)
    
    avr_train_score.append(np.mean(train_score))
    avr_valid_score.append(np.mean(valid_score))
```

```python
sns.lineplot(y=avr_train_score,x=np.arange(2,150,5),c='red',label='train')
sns.lineplot(y=avr_valid_score,x=np.arange(2,150,5),c='blue',label='valid')
plt.ylabel('accuracy_score')
plt.xlabel('min_samples_leaf')
```

```python
#max accuracy_score is 0.81 with min_samples_leaf at 42. 
avr_valid_score
```

```python
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier
from catboost import CatBoostClassifier
import time
```

```python
n = 150
models = {
    'XGBoost': XGBClassifier(random_state = 78, n_estimators=n,max_depth=3,subsample=0.8,eval_metric='auc',colsample_bytree=0.5),
    'LightGBM': LGBMClassifier(verbose=-100,random_state = 2, n_estimators=n, max_depth=3),
    'CatBoost': CatBoostClassifier(verbose=0,random_state=44, n_estimators=n,max_depth=3)
    }
```

```python
result = {}
for name, model in models.items():
    time_start = time.time()
    model.fit(df[features],df[target])
    time_taken = time.time() - time_start
    pred_train = model.predict(df[features])
    pred_test = model.predict(df_test[features])
    train_score = accuracy_score(df[target],pred_train)
    test_score = accuracy_score(df_test[target],pred_test)
    
    result[name] = {'train_score': train_score,
                   'test_score': test_score,
                   'time': time_taken}
```
